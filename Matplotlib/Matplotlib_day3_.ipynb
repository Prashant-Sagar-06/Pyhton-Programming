```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Question 1:- Plotting a Linear Function
x = np.linspace(-10, 10, 100)
y = 2*x + 3

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Function: y = 2x + 3')
plt.grid(True)
plt.show()
```

```python
#Question 2 :- Plotting Multiple Functions
x = np.linspace(0, 10, 100)
y1 = x**2
y2 = np.sqrt(x)

plt.plot(x, y1, label='y = x^2', linestyle='-', color='blue')
plt.plot(x, y2, label='y = sqrt(x)', linestyle='--', color='red')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of y = x^2 and y = sqrt(x)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Question 3:- Plotting Trigonometric Functions
x = np.linspace(0, 2*np.pi, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)
y_tan = np.tan(x)

plt.plot(x, y_sin, label='y = sin(x)')
plt.plot(x, y_cos, label='y = cos(x)')
plt.plot(x, y_tan, label='y = tan(x)')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Trigonometric Functions')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Question 4:- Plotting Exponential Function
x = np.linspace(-5, 5, 100)
y = np.exp(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of y = e^x')
plt.axvline(0)
plt.grid(True)
plt.show()
```

```python
#Question 5:- Plotting Piecewise  Function
x = np.linspace(-2, 2, 100)
y = np.where(x < 0, -1, 1)

plt.plot(x, y,linestyle='--', color='red')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Piecewise Function')
plt.grid(True)
plt.show()
```

```python
#Question 6:- Plotting a Parametric Function
t = np.linspace(0, 2*np.pi, 100)
x = np.cos(t)
y = np.sin(t)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Parametric Equations: x = cos(t), y = sin(t)')
plt.grid(True)
plt.show()
```

```python
#Question 7:-. Plotting Logarithmic Functions
x = np.linspace(0.1, 6, 100)
y1 = np.log(x)
y2 = np.log10(x)
plt.plot(x, y1, label='y = ln(x)')
plt.plot(x, y2, label='y = log10(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Logarithmic Functions')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Question 8:- Plotting Step Function 
x = np.linspace(-5, 5, 100)
y = np.where(x < 0, 0, 1)

plt.step(x, y)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Step Function')
plt.grid(True)
plt.show()
```

```python
#Question 9:- Plotting a Polynomial Function
x = np.linspace(-3, 3, 100)
y = x**3 - 2*x**2 + x - 1
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of y = x^3 - 2x^2 + x - 1')
plt.grid(True)
plt.show()
```

```python
#Question 10:- Plotting Piecewise Linear Function 
x = np.linspace(-2, 2, 100)
y = np.where(x <= 0, -x, x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Piecewise Linear Function')
plt.grid(True)
plt.show()
```
